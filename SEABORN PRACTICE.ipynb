```python
df=pd.read_csv('Who_is_responsible_for_global_warming.csv')
df.head()
```

```python
df=df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name')
df.head()
```

```python
sns.heatmap(df)
```

```python
plt.figure(figsize=(14,9))
sns.heatmap(df, vmin=0,  vmax=21,  cmap='coolwarm' ,annot=True,fmt='0.3f' xticklabel=np.arange(0,15))


#cmap = for color
#annot= TO SHOW values in each box
```

```python
plt.figure(figsize=(16,9))
annot_kws={'fontsize': 10,'fontstyle':'italic',
                                                        'color':'k',
                                                        'alpha': 0.8,
                                                        'backgroundcolor':'w',
                                                        'verticalalignment':'center',
                                                        'rotation':'vertical'}
                                                        


hm=sns.heatmap(df, cmap='coolwarm',  annot=True, annot_kws=annot_kws,  linewidth=3,  linecolor='k', cbar=False) 


hm.set(title='heatmap of who is responsible for global warming',
      xlabel='co2 emiisions per year',
      ylabel='countries names')

sns.set(font_scale=0.8)
#xticklabels=False to hide x axis

```

```python
plt.figure(figsize=(16,9))

cbar_kws={'orientation':'horizontal',
         'shrink':1,
         'extend':'min',
         'extendfrac':0.08,
         'ticks':np.arange(0,22),
         'drawedges':True}

sns.heatmap(df, cmap='coolwarm', cbar_kws=cbar_kws,  annot=True)
```

```python
df.corr()
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(df.corr(), annot=True, linewidth=3)

ax.tick_params(size=10, color='w', labelsize=10, labelcolor='k')
plt.title('heatmap of who is responsible for global warming',  fontsize=17)
```
