# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random 
```

```python
ml_student_age= np.random.randint(18,45,100)
py_student_age=np.random.randint(15,40,100)
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
bin1=[15,20,25,30,35,40,45]
#bin2=[15,20,25,30,35,40]
```

```python
plt.hist(ml_student_age,  bin1, rwidth=0.8,  histtype='bar', align='mid',  
         orientation='horizontal',  color='g', label='ml student')


#rwidth= by default it is 1
#align= it can be left , mid, right
#orientation= horizontal, vertical
#label=  always give plt.legend if you want to show your label on graph


plt.legend(loc=4)  #
plt.xlabel('ml_student_age')
plt.ylabel('no.student age')
plt.title('ml student histogram')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')

plt.figure(figsize=(15,10))
plt.hist([ml_student_age, py_student_age],  bin1, rwidth=0.8,  histtype='bar', align='mid',  
         orientation='vertical',  color=['g','y'], label=['ml student','py student'])

#plt.hist(py_student_age,  bin2, rwidth=0.8,  histtype='bar', align='mid',  orientation='vertical',  color='b', label='py student')


#rwidth= by default it is 1
#align= it can be left , mid, right
#orientation= horizontal, vertical
#label=  always give plt.legend if you want to show your label on graph


plt.legend(loc=4)  #
plt.xlabel('ml_student_age',color='k', size=20)
plt.ylabel('no.student age',color='k', size=20)
plt.title('ml student histogram')


plt.grid(color='k', linewidth=1, linestyle="--" )
plt.show()
```

# PIE CHART

```python
parts =            [7,         2,               20,             3]
activity =        ['sale',   'manufacture',   'manpower',   'maintaince']
colours_we_want = ['green',  '#66b3ff',      '#99ff99',    '#ffcc99']# these are  RGB hex colour codes for colours

fig = plt.figure(figsize = (5,5)) # (width, height)

plt.pie( parts ,# data
        labels = activity ,
        explode = (0.1, 0.2, 0, 0),   # explode slices out a particular slice , brings it out , here 3rd slice is tsken out.     
        colors = colours_we_want ,      
        
        autopct = '%0.3f%%',     # autopct enables you to display the percent value using Python string formatting. #%0.3f
                                 #... 2f' , then for each pie wedge, the format string is '%. 2f' autopct shows percent weight of slice ,  we have to learn the string code
        startangle = 90 ,       # startangle makes the starting aligned at 90 degree and then moves counter clockwise
        shadow = True ,          # shadow displays the shadow of slices
        radius =  1.2,           # for radius of the circle
        counterclock=True ,      # labeling done counter clockwise
        textprops={'fontsize': 15,
                  'color':'black'} );         
```

```python
plt.figure(figsize=(6,4))

parts=[ 25,  15,  40, 18,  12 ]
colorr= ['g', 'b', 'r' ,'m', 'y']
sub=['english', 'maths', 'ds', 'python','R']
wedgeprops={'linewidth': 4, 'width':1, 'edgecolor':'k'}


plt.pie(parts,
       colors= colorr,
       labels= sub,
       labeldistance=1.5,
       explode= [0.2,0,0,0.1,0],
       autopct='%0.1f%%',
       shadow=True,
       radius=1.1,
       startangle=90,
       pctdistance= 0.8,
       counterclock=True,
       wedgeprops=wedgeprops,
       center=(2,3),
       
       rotatelabels= True,
       textprops={'fontsize': 15,
                 'color': 'black'}
       )
plt.legend(loc=10)
plt.show()
```

```python
plt.figure(figsize=(6,3))

colors=['w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r']
data=np.ones(20) 
#labels =[1.0,1.0,1.0........]

plt.pie([1], colors='k', radius=2.05)
plt.pie(data, colors=colors, radius=2.0)
plt.show()
```

```python
plt.figure(figsize=(6,3))

colors=['w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r']
data=np.ones(20) 
#labels =[1.0,1.0,1.0........]

plt.pie([1], colors='k', radius=2.05)
plt.pie(data, colors=colors, radius=2.0)

plt.pie([1], colors='g', radius=1.85)       #using 1 because i don't want slices 
plt.pie([1], colors='y', radius= 1.70)
plt.pie([1], colors='b', radius= 1.50)
plt.pie([1], colors='m',  radius=1.2)

plt.pie(data,  colors='b',  radius= 0.7)
plt.pie(data, colors= colors, radius= 0.68)

plt.pie(data, colors='w',  radius= 0.5)
plt.pie(data,  colors= 'k',  radius=0.3)



plt.savefig("my_pie_chart.png",
           dpi=100,
           quality=99,
           facecolor='w',
            orientation='potrait' 
           )     
#by default it will be saved in .png format (portable network graphic) it supports loseless data compression

plt.show()

```
